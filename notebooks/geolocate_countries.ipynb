```python
def get_lat_long(query):
    results = geocoder.geocode(query)   
    lat = results[0]['geometry']['lat']  if results else None
    long = results[0]['geometry']['lng'] if results else None
    return query, lat, long
```

```python
import pandas as pd
world_bank = pd.read_csv("../data/World_Bank_Clean.csv")
world_bank.head(2)
```

```python
countries = world_bank.Country.unique()
```

```python
from unpythonic import begin
country_lat_longs = [begin(print('Processing {0}'.format(c), get_lat_long(c)), get_lat_long(c)) for c in countries]
country_lat_longs
```

```python
# Only missing lat,long was Macedonia, FYR, which was located manually using 
# https://nominatim.openstreetmap.org/details.php?place_id=198021110
lats = {**{c:lat 
           for c, lat, long in country_lat_longs},
        'Macedonia, FYR':41.6171214}
longs = {**{c:long 
            for c, long, long in country_lat_longs},
        'Macedonia, FYR':21.7168387}
```

```python
from dfply import *
world_bank_after_1990_w_lat_long = (world_bank
                                    >> filter_by(X.Year > 1990)
                                    >> mutate(Latitude = X.Country.map(lats),
                                              Longitude = X.Country.map(longs))
                                    >> arrange(X.Scandinavia)
)
world_bank_after_1990_w_lat_long
```

```python
url = '../data/world_bank_filtered_w_lat_long.json'
world_bank_after_1990_w_lat_long.to_json(url, orient='records')
```
